```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
LL= q1-(IQR*1.5)                 
UL = q3+(IQR * 1.5)
lower= df<LL
upper=df>UL
out=df[~(lower | upper).any(axis=1)]                                         #remove outliers
df=out
df.shape
```
