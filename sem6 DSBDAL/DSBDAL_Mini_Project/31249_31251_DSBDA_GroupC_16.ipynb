##### As many null values are present for first and second dose administered columns the operations on them will be inaccurate.

```python
avg_firstdose = data1["First Dose Administered"].astype("float").mean(axis = 0)
data1["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data1
```

```python
avg_seconddose = data1["Second Dose Administered"].astype("float").mean(axis = 0)
data1["Second Dose Administered"].fillna(value = avg_seconddose, inplace=True)
data1
```

### 2)Number of persons state wise vaccinated for first dose in India

```python
first_dose = data1.groupby('State')[['First Dose Administered']].sum()
first_dose.drop(index="India",axis=0)
```

```python
data2['State/UnionTerritory']=data2['State/UnionTerritory'].replace(['Bihar****','Himanchal Pradesh','Karanataka','Madhya Pradesh***','Maharashtra***','Telengana'],['Bihar','Himachal Pradesh','Karnataka','Madhya Pradesh','Maharashtra','Telangana'])
```

```python
cured_cases = data2.groupby('State/UnionTerritory')[['Cured']].sum()
cured_cases
```

```python
confirmed_cases = data2.groupby('State/UnionTerritory')[['Confirmed']].sum()
confirmed_cases
```

```python
death_cases = data2.groupby('State/UnionTerritory')[['Deaths']].sum()
death_cases
```

```python
data2['active_cases']=data2['Confirmed']-data2['Cured']-data2['Deaths']
active_cases=data2.groupby('State/UnionTerritory')[['active_cases']].sum()
active_cases
```

```python
top_5_active_cases = data2.groupby(by = 'State/UnionTerritory').max()[['active_cases']].sort_values(by = ['active_cases'], ascending= False).reset_index()
top_5_active_cases
```

```python
import seaborn as sns
ax = sns.barplot(data= top_5_active_cases.iloc[:5], y = 'active_cases', x= 'State/UnionTerritory', linewidth = 2, edgecolor = 'red')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (13,6))
ax = sns.lineplot(data = data2[data2['State/UnionTerritory'].isin(['Maharasthra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x = 'Date', y = 'active_cases', hue = 'State/UnionTerritory')
ax.set_title('Top 5 most affected States', size = 10)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (13,6))
plt.title("Top 5 vacccinated States", size =20)
max_vac = data1.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
max_vac = max_vac.sort_values('Total Individuals Vaccinated', ascending = False)[1:6]
max_vac
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac['Total Individuals Vaccinated'], x = max_vac.index, linewidth = 2, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show
```
